## Handling Missing values

```python
# df['location'].mode()[0]

# df['location']=df['location'].fillna('Whitefield')
df['location']=df['location'].fillna(df['location'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['size'].unique()
```

```python
df['size']=df['size'].fillna(df['size'].mode()[0])
```

```python
df['bath'].value_counts()
```

```python
df['bath']=df['bath'].fillna(2.0)
```

```python
df.isnull().sum()
```

```python
df['balcony'].unique()
```

```python
df['balcony']=df['balcony'].fillna(0.)
```

## Treating Inconsistent Data

```python
df['location'].value_counts()
```

```python
df['location']=df['location'].apply(lambda x:x.strip())
```

```python
loc=df['location'].value_counts()
loc_less_than_10=loc[loc<=10]
df['location']=df['location'].apply(lambda x: 'others' if x in loc_less_than_10 else x)
```

```python
df['location'].value_counts()
```

```python
df['size'].unique()
```

```python
out=[]
for i in df['size']:
    out.append(int(i.split()[0]))
df['size']=out
df.head()
```

```python
df.info()
```

```python
df['total_sqft'].unique()
```

```python
def clean_sqft(sqft):
    l=sqft.split('-')
    if len(l)==2:
        return (float(l[0])+float(l[1]))/2
    try:
        return float(l[0])
    except:
        return None
```

```python
df['total_sqft']=df['total_sqft'].apply(clean_sqft)
```

```python
df['total_sqft'].unique()
```

```python
df['total_sqft'].dtype
```

```python
df['bath']=df['bath'].astype(int)
```

```python
df['balcony']=df['balcony'].astype(int)
```

```python
df.info()
```

```python
df['total_sqft'].isnull().sum()
```

```python
df.dropna()
```

## Feature Engineering

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
df.head()
```

## Handling outliers

```python
df.describe()
```

```python
df['size'].value_counts()
```

```python
#Handling the outliers in size column
df=df[df['size']<=6]
df['size'].unique()
```

```python
#Handling the outliers in bath column
df=df[df['bath']<df['size']+2]
df['bath'].unique()
```

```python
sns.boxplot(data=df,y='price_per_sqft')
```

```python
q1=df['price_per_sqft'].quantile(0.25)
q3=df['price_per_sqft'].quantile(0.75)

IQR=q3-q1

lower=q1-1.5*IQR
upper=q3+1.5*IQR

df=df[(df['price_per_sqft']>=lower)&(df['price_per_sqft']<=upper)]
```

```python
sns.boxplot(data=df,y='price_per_sqft')
```

```python
df=df.reset_index(drop=True)
```

## Machine Learning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
encoder=LabelEncoder()
df['encoded_loc']=encoder.fit_transform(df['location'])
df
```

```python
df1=df
```

```python
df=df.drop(['location','price_per_sqft'],axis=1)
df
```

```python
X=df.drop(['price'],axis=1)
y=df.price
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestRegressor(random_state=42)
params={'n_estimators':[100,150,200,250,300],
        'max_depth':[3,4,5,6,7]}
grid=GridSearchCV(estimator=model,param_grid=params,cv=5)
grid.fit(Xtrain,ytrain)
print('Best parameters: ',grid.best_params_)
print('Best score: ',grid.best_score_)
```

```python
ypred=grid.predict(Xtest)
ypred
```

```python
print('Training efficiency',grid.score(Xtrain,ytrain))
print('Testing efficiency',grid.score(Xtest,ytest))
```

```python
print('R2 score',r2_score(ytest,ypred))
print('Mean Absolute Error',mean_absolute_error(ytest,ypred))
```
